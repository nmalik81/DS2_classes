```python
import numpy as np
from random import sample
import numpy.random as nr
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

def sampling_distribution(p,n,m):
    smean_dist = []
    for i in range(0,m):
        smean_dist.append(np.mean(np.array(sample(list(p),n))))
    return smean_dist
    
```

```python
a=nr.exponential(10,50000)
b=nr.uniform(0,10,50000)
c=nr.normal(0,5,50000)
dist_1 = sampling_distribution(a,2000,1000)
dist_2 = sampling_distribution(b,2000,1000)
dist_3 = sampling_distribution(c,2000,1000)

sns.set(style="white",palette="muted",color_codes = True)
f,axes=plt.subplots(3,2,figsize=(13,11))
sns.despine(left=True)

sns.distplot(a,color="b",ax=axes[0,1],axlabel="Sampling distribution for exponential")
sns.distplot(a,color="b",ax=axes[1,1],axlabel="Sampling distribution for uniform")
sns.distplot(a,color="b",ax=axes[1,0],axlabel="Sampling distribution for normal")
```
