```python
population=np.random.randint(10,20,1000)
np.random.seed(1)
estimates=[]
for x in range(200):
    sample=np.random.choice(a=population,size=100)
    estimates.append(sample.mean())
```

```python
import random
from collections import Counter

def toss_coin(n):
    tosses=[]
    for i in range(n):
        if random.getrandbits(1)==0:
            tosses.append('H')
        else:
            tosses.append('T')
    return tosses

def get_probability(tosses,result):
    counter = Counter(tosses)
    return counter[result]/len(tosses)

tosses = toss_coin(10)
print(tosses)
            
    
```
